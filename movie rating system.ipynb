```python
movies = []
def add_movie():
    name = str(input("enter the name of movie"))
    rating = float(input("enter the rating of movie"))
    movie = {
        "name":name,
        "rating":rating
    }
    movies.append(movie)
def display_movies():
    for m in movies:
        print(f"name of movie:{m['name']} and rating of movie:{m['rating']}")
def search_movie():
    name = str(input("enter the name of movie"))
    for m in movies:
        if name == m["name"]:
            print(f"name:{name},rating:{m["rating"]}")
        else:
            print("name of movie not in moies")
def highest_movie():
    highest_rating = 0
    name = ""
    for m in movies:
        if m['rating'] > highest_rating:
            highest_rating = m['rating']
            name = m['name']
            print(f"the highest movie of {name} is {highest_rating}")
def delete_movie():
    name = str(input("enter the name"))
    for m in movies:
        if name == m['name']:
            movies.remove(m)
        else:
            print("movie not found")
while True:
    print("1-add_movie")
    print("2-display_movies")
    print("3-search_movie")
    print("4-highest_movie")
    print("5-delete_movie")
    print("6-exit")
    choice = input("enter the choice")
    if choice == "1":
        print(add_movie())
    elif choice == "2":
        print(display_movies())
    elif choice == "3":
        print(search_movie())
    elif choice == "4":
        print(highest_movie())
    elif choice == "5":
        print(delete_movie())
    elif choice == "6":
        print("program end")
        break
    else:
        print("enter true choice")
        
```
