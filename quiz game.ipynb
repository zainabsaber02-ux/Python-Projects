```python
questions = [
    {
        "question":"what is the capital of egypt",
        "answer":"cairo"
    },
    {
        "question":"how many days are there in a week",
        "answer":"7"
    },
    {
        "question":"what colour is the sky",
        "answer":"blue"
    },
    {
        "question":"what is 3+5?",
        "answer":"8"
    }
]
score = 0
for q in questions:
    answer_user = input(f"{q["question"]}").lower()
    if answer_user == q["answer"].lower():
        print("your answer is true")
        score += 1
    else:
        print("answer is not true")
        
print(f"you gain {score} score")
if score == len(questions):
    print("excellant")
elif score >= 2:
    print("good")
else:
    print("try again")
```
