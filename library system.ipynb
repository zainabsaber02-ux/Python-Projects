```python
library = {}

#اضافه عناصر للقاموس 

library["computer science"] = {"author": "john","statues": "available"}
library["data stracture"] = {"author": "mark","statues": "available"}

#استعراض جميع الكتب 
for title,info in library.items():
    print(f"{title} المؤلف هو {info["author"]},الحاله هي {info["statues"]}")
    
#استعاره كتاب 
book_to_borrow = "computer science"
if book_to_borrow in library:
    if library[book_to_borrow]["statues"] == "available":
        library[book_to_borrow]["statues"] == "borrowed"
        print(f"الكتاب{book_to_borrow} انت استعرته ")
    else:
        print(f"الكتاب{book_to_borrow} غير موجود حاليا")
else:
    print(f"الكتاب {book_to_borrow} غير موجود في المكتبه")
        
#ارجاع كتاب 
book_to_return = "computer science"
if book_to_return in library :
    if library[book_to_return]["statues"] == "borrowed":
        library[book_to_return]["statues"] == "available"
        print(f"لقد تم ارجاع كتاب {book_to_return}")
    else:
        print(f"الكتاب{book_to_return}غير مستعار")
else:
    print(f"الكتاب{book_to_return} غير موجود في المكتبه ")
    
#استعراض الكتب بعد اجراء العمليات عليها 
for title,info in library.items():
   print(f"{title} المؤلف هو {info["author"]} الحاله هي {info["statues"]}")
    
del library["computer science"]
for title,info in library.items():
    print(f"{title} المؤلف هو {info["author"]} الحاله هي {info["statues"]}")
```
