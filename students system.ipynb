```python
students = []
def add_student():
    name = str(input("enter the name"))
    marks = {}
    
    subjects = ["math","english","science"]
    for sub in subjects:
        mark = float(input("enter the mark"))
        marks[sub] = mark
        
    student = {
       "name":name,
       "marks":marks
    }
    students.append(student)
        
def show_students():
    
    if not students:
        print("name not found")
        return
    for s in students:
        marks = s["marks"]
        aveg = sum(marks.values())/len(marks)
        status = "pass" if aveg > 50 else "fail"
        print(f"name:{s["name"]},marks:{marks},average:{aveg},status:{status}")
        
def search_student():
    name = str(input("ente the name"))
    for s in students:
       if name == s["name"]:
          print(f"name:{name},marks:{s["marks"]}")
    else:
        print("name not found")
            
def delete_student():
    name = str(input("enter the name"))
    if name not in students:
        print("student not found")
    else:
        for s in students:
           students.remove(s[name])
        
while True:
    print("1-add_student")
    print("2-show_students")
    print("3-search_student")
    print("4-delete-student")
    print("5-exit")
    choice = input("enter your choice")
    if choice == "1":
        add_student()
    elif choice == "2":
        show_students()
    elif choice == "3":
        search_student()
    elif choice == "4":
       delete_student()
    elif choice == "5":
        print("programm end")
        break
    else:
        print("invaild")
```
